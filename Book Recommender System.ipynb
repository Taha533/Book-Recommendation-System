### Popularity based recommender System:

```python
ratings_with_name = ratings.merge(books, on = "ISBN")
ratings_with_name.head()
```

```python
ratings_with_name.groupby("Book-Title").count()
```

```python
num_rating_df = ratings_with_name.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_rating_df.rename(columns={"Book-Rating": "num_ratings"}, inplace = True)
num_rating_df
```

```python
num_rating_df["num_ratings"].max()
```

```python
avg_rating_df = ratings_with_name.groupby("Book-Title").mean()["Book-Rating"].reset_index()
avg_rating_df.rename(columns={"Book-Rating": "avg_ratings"}, inplace = True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on = "Book-Title")
popular_df
```

```python
popular_df = popular_df[popular_df["num_ratings"] >= 250].sort_values("avg_ratings", ascending = False).head(50)
popular_df
```

>W e want to get the publisher name also

```python
popular_df = popular_df.merge(books, on = "Book-Title").drop_duplicates("Book-Title")[["Book-Title", "Book-Author","Image-URL-M","num_ratings","avg_ratings"]]
popular_df
```

```python
popular_df["Image-URL-M"].values[0]
```

>Collaborative Filtering:

```python
ratings_with_name.head()
```

```python
x = ratings_with_name.groupby("User-ID").count()['Book-Rating']>200
regular_users = x[x].index
regular_users
```

```python
ratings_with_name
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(regular_users)]
filtered_rating
```

```python
y = filtered_rating.groupby("Book-Title").count()['Book-Rating']>50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
final_ratings["Image-URL-M"].values[0]
```

```python
pt = final_ratings.pivot_table(index = "Book-Title",columns = "User-ID", values = "Book-Rating")
pt
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt)
similarity_scores
```

```python
cosine_similarity(pt).shape
```

```python
#0th index book will no be considered because it is the book itself
sorted(list(enumerate(similarity_scores[0])),key = lambda x:x[1],reverse = True)[1:6]
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key = lambda x:x[1],reverse = True)[1:6]
    data =[]
    for i in similar_items:
        #print(pt.index[i[0]])
        item = []
        tmp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(tmp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(tmp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(tmp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
        
    return data
        
```

```python
recommend("Harry Potter and the Chamber of Secrets (Book 2)")
```
